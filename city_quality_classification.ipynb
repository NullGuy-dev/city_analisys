importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

getting dataset

```python
data = pd.read_csv("/content/drive/MyDrive/city_analisys_data.csv")
data.head()
```

preprocessing data

```python
colm = ["is_people_happy", "count_of_schools", "count_of_work_places", "average_salary_of_resident"]
label = ["city_rating"]

scaler = StandardScaler()
scaler = scaler.fit(data[colm])
data[colm] = scaler.transform(data[colm])
data.head()
```

splitting data

```python
X, y = train_test_split(data, test_size=0.2, shuffle=True)
X_train, X_test, y_train, y_test = X[colm], y[colm], X[label], y[label]
```

training random forest for classification | Random Forest

```python
rand_forest = RandomForestClassifier(max_depth=5, random_state=42)
rand_forest.fit(X_train, y_train)
```

training random forest for classification | KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

testing model

```python
from sklearn.metrics import f1_score, accuracy_score

y_pred = rand_forest.predict(X_test) # random forest is choosed for example
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("F1 score: ", f1_score(y_test, y_pred, average="micro"))
```
